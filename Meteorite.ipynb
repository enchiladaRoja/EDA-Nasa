Importamos las librerias que necesitaremos

```python
import pandas as pd
import numpy as np
```

Asignamos el csv del que deseamos obtener los valores a la variable URL


```python

url = "C:/Users/Fab_V/Documents/EDAnasa/Meteorite_Landings.csv"
```

Creamos un dataframe con la funcion de leer csv, al cual le pasamos la url con la ubicacion del csv a leer, utilizamos head() para comprobar que el dataframe ya esta poblado con los datos del csv

```python
df = pd.read_csv(url)
df.head()
      
```

Analizamos el dataframe, eliminamos las columnas que no necesitamos con drop

```python
df.drop(['recclass','GeoLocation'], axis = 1, inplace=True)
```

Comprobamos que las columnas se hayan eliminado con head()

```python
df.head()
```

Utilizamos info() para ver que tipo de datos tienen las columnas

```python
df.info()
```

Creamos una nueva columna llamada year2, donde tomamos los primeros 10 caracteres de la columna year. Como vimos con info(), la columna year es de tipo object y no de tipo datetime, solo necesitamos la fecha y no el timestamp.

```python
df['year2'] = df['year'].str[:10]
df.head()
```

Eliminamos la columna year, ya que no la necesitaremos

```python
df.drop(['year'], axis = 1, inplace=True)
```

Comprobamos que la columna se haya eliminado y que la nueva columna este en el dataframe

```python
df.head()
```

Renombramos la columna 'year2' como 'year', comprobamos que se confirme el cambio

```python
df.rename(columns={'year2':'year'}, inplace=True)
df.head()
```

Buscamos datos perdidos en el dataframe y los cambiamos por NaN de la libreria Numpy

```python
df.replace("?", np.nan, inplace = True)
df.head(5)
```

Creamos un dataframe con los valores nullos del dataframe con la isnull(), esto nos regresara un valos booleano en cada posicion del dataframe, siendo "True" cuando encuentre valores nulos y "False" cuando encuentre informacion.

```python
missing_data = df.isnull()
missing_data.head(5)
```

Recorremos el dataframe creado con un for y obtenemos su informacion, mostrandonos el numero de valores True and False, con lo cual sabremos donde tenemos informacion perdida

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")    
```

Con la informacion obtenida, analizamos si podemos eliminar filas que contienen valores no encontrados, de decidirlo asi, despues resetamos el index

```python
df.dropna(subset=["reclong"], axis=0, inplace=True)
df.reset_index(drop=True, inplace=True)
```

Una de las columnas tiene varias filas de valores perdidos que no deseamos eliminar, asi que cambiaremos los valores por la media de esa columna. Obtenemos la media y a asignamos a una variable, imprimimos para conocer el valor.

```python
avg_mass=df['mass (g)'].astype('float').mean(axis=0)
print("Average of mass:", avg_mass)
```

Ya con el valor de la media de esa columna, cambiamos los valores perdidos de las finas antes encontradas con este valor.

```python
df["mass (g)"].replace(np.nan, avg_mass, inplace=True)
```

```python
df.head()
```

Cambiamos el nombre de esta columna por uno que nos parece mas manejable

```python
df.rename(columns={'mass (g)':'mass_grams'}, inplace=True)
df.head()
```
